```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
market = pd.read_csv("/content/drive/MyDrive/UTS_MPML/transaction_data.csv")
market
```

```python
market.info()
```

```python
market.isnull().sum()
```

```python
market.dropna()
```

```python
market=market[market['NumberOfItemsPurchased']>=0]
```

```python
transaction=(market.groupby(['TransactionId','ItemDescription'])['NumberOfItemsPurchased'].sum().unstack().reset_index().fillna(0).set_index('TransactionId'))
transaction
```

```python
def encode_units(x):
  if x<=0:
    return 0
  if x>= 1:
    return 1

transaction_encode = transaction.applymap(encode_units)
```

```python
transaction_encode
```

```python
transaction_filter = transaction_encode[(transaction_encode > 0).sum(axis=1) >= 2]
transaction_filter
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
transaction_itemset = apriori(transaction_filter, min_support=0.03,
                                 use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)

transaction_itemset['length'] = transaction_itemset['itemsets'].apply(lambda x: len(x))

transaction_itemset
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(transaction_itemset, metric='lift',
                  min_threshold=1).sort_values('lift', ascending=False).reset_index(drop=True)
```
